```python
# Q41: Create a histogram for student exam scores (in marks).

# Import matplotlib library for plotting graphs
import matplotlib.pyplot as plt

# Marks obtained by students (out of 100)
marks = [45, 50, 55, 60, 62, 65, 68, 70, 72, 75, 78, 80, 83, 85, 90]

# Create histogram with 5 groups
plt.hist(marks, bins=5, color='blue', edgecolor='black')

# Label the axes
plt.xlabel("Marks (out of 100)")
plt.ylabel("Number of Students")

# Title of the graph
plt.title("Student Exam Score Distribution")

# Display the graph
plt.show()
```

```python
# Q42: Create a histogram for employee salary distribution (₹ per month).

import matplotlib.pyplot as plt  # Import library

# Monthly salary data of employees (in rupees)
salary = [25000, 28000, 30000, 32000, 35000, 38000, 40000, 42000, 45000, 48000]

# Plot histogram
plt.hist(salary, bins=5, color='green', edgecolor='black')

# Axis labels
plt.xlabel("Monthly Salary (₹)")
plt.ylabel("Number of Employees")

# Graph title
plt.title("Employee Salary Distribution")

plt.show()  # Show graph
```
