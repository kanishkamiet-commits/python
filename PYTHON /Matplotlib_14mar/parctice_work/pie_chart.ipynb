```python
# Import matplotlib library
import matplotlib.pyplot as plt

# Data for mobile brands
brands = ['Samsung', 'Apple', 'Xiaomi', 'Oppo', 'Others']
share = [30, 25, 20, 15, 10]

# Colors for each brand
colors = ['blue', 'grey', 'orange', 'green', 'red']

# Create pie chart
plt.pie(share, labels=brands, colors=colors, autopct='%1.1f%%')

# Title of chart
plt.title("Mobile Market Share")

# Display chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Courses and number of students
courses = ['BCA', 'BBA', 'B.Tech', 'MBA', 'MCA']
students = [40, 30, 50, 20, 25]

# Different colors
colors = ['purple', 'yellow', 'cyan', 'orange', 'pink']

# Draw pie chart
plt.pie(students, labels=courses, colors=colors, autopct='%1.1f%%')

# Title of chart
plt.title("Student Course Distribution")
# Display chart
plt.show()
```
